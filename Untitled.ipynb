```python
add = lambda a , b : a+b
print(add(5 , 6))
```

```python
multiply = lambda a , b , c : a*b+c
print(multiply(5,6,5))
```

```python
even = lambda x: x % 2 == 0

print(even(5))
```

```python
numbers = [1, 2, 3, 4, 5, 6]

even_numbers = list(filter(lambda x: x % 2==0,numbers))

print(even_numbers)
```

```python
students = [
    ("John", 85, "a"),
    ("Mike", 72, "e"),
    ("Anna", 95, "c")
]

sorted_students = sorted(students,key=lambda x: x[0],reverse=True)

print(sorted_students)
```

```python
grade = lambda x: "pass" if x >= 60 else "fail"

print(grade(75))
print(grade(40))
```

```python
status = lambda age: ("Child" if age < 18 else "Adult" if age < 60  else "Senior")

print(status(10))
print(status(30))
print(status(70))
```

```python
from functools import reduce

numbers = [1, 2, 3, 4]

total = reduce(lambda x, y: x + y, numbers
)

print(total)
```

```python
import pandas as pd

df = pd.DataFrame({
    "salary": [3000, 7000, 12000]
})

df
```

```python
df["category"] = df["salary"].apply(lambda x: "Low" if x < 5000 else "Medium" if x < 10000  else "High")

print(df)
```

```python
employees = [
    {"name": "John", "salary": 5000},
    {"name": "Anna", "salary": 9000},
    {"name": "Mike", "salary": 7000}
]

sorted_emp = sorted(employees,key=lambda x: x["name"], reverse=True)

print(sorted_emp)
```

```python
import pandas as pd
import numpy as np

df = pd.DataFrame({
    "delivery_time": [20, 45, 90, 120]
})

df["delivery_status"] = df["delivery_time"].apply(lambda x:"Fast" if x < 30 else "Medium" if x < 90 else "Slow")

print(df)
```
