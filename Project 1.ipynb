```python
#Which three dates have the highest traffic?(find this programmatically). 
from datetime import datetime
df['Pickup_date']=pd.to_datetime(df['Pickup_date']).dt.date #converting to date format and removing time stamp
df2=df['Pickup_date'].value_counts().head(3)
print(df2)
print("the three dates which have higest traffic are 2015-06-17,2015-05-17 & 2015-05-19")
```

```python
#Which three dates have the highest traffic? Which days (Mon, Tue?) were these (find this programmatically). 
from datetime import datetime
df['Pickup_date']=pd.to_datetime(df['Pickup_date']) #converting to date format and removing time stamp
df['Day']=df['Pickup_date'].dt.day_name()
df2=df.groupby(['Pickup_date','Day']).size().reset_index(name='count')
print(df2)
df3=df2.sort_values(by='count',ascending=False)
print(df3)
print("the three dates which have higest traffic are 2015-06-17=Wednesday,2015-05-17=Sunday & 2015-05-19=Tuesday")
```

```python
#Summarize data per day (eliminate time) and plot the number of pickups per day. If days are too many, summarize per week. 
fname='uber-raw-data-janjune-15.csv'
df=pd.read_csv(fname)
df['Pickup_date']=pd.to_datetime(df['Pickup_date']).dt.date
x=df['Pickup_date'].value_counts()
print(x)
plt.plot(x,color="green",label='Pickups per day')
plt.suptitle('Uber pickup data',fontsize=18,color='black') 
plt.xlabel('Date')
plt.ylabel('Number of pickups')
plt.legend()
#ref:Data 605 assignment2 dt:09-06-2022
```

```python
#Is there an Affiliated_base_num which has some interesting statistics (most people visit it on a certain day of the week?)
df['Pickup_date']=pd.to_datetime(df['Pickup_date']) #converting to date format and removing time stamp
df['Day']=df['Pickup_date'].dt.day_name()#converting date into day of the week
df1=df.groupby(['Affiliated_base_num','Day']).size().reset_index(name='count')
print(df1)
print(df1.max())
print('Affiliagtd base number B02847 is the highest visited base for the data.')
print('Wednesday is the day of the week in which most people visited for the data.')
```
