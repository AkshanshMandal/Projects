```python
# Import libraries and data

import pandas as pd
%matplotlib inline

body_data = pd.read_csv('class_survey_body_measurements.csv')
col_names = ['timestamp', 'sex', 'handspan', 'height', 
             'shoe_size','hair_color', 'mother_height', 'mother_shoe_size',
             'mother_hair_color', 'father_height', 'father_shoe_size',
             'father_hair_color','athlete','shoulder_width','skull_circum']
body_data.columns = col_names
body_data['handspan'][37]
```

```python
body_data.info()
```

<span  style="color:purple; font-size:16px">Creating a scatterplot of `height` (y-axis) versus `handspan` (x-axis).</span>

```python
# Answer
body_data.plot(kind= 'scatter',
              x = 'handspan',
              y = 'height',
              title = "Scatter plot for Handspan VS Height",
              xlabel = 'Handspan',
              ylabel = 'Height',
              legend = True,
              figsize = (10,8))
```

## As per the plot there are some outliers between handspan 50 to 70 whereas all other handspan are on either handspan nearer 10 or below handspan 10 which suggests that these outliers could represent unusual values in the dataset and we need to investigate whether these outliers are geniune or errors.

<span  style="color:purple; font-size:16px">Setting observations where `handspan` is greater than 30 or less than 4 to be missing using `pd.np.nan`. </span>

```python
# Answer
import numpy as np
filt = (body_data['handspan'] > 30) | (body_data['handspan'] < 4)
extract_index = body_data.loc[filt,'handspan'].index
body_data.loc[extract_index,'handspan'] = pd.np.nan
body_data.sample(10)
```

<span  style="color:purple; font-size:16px">Importing the Seaborn library and the `regplot()` function to fit a line of best fit through the data to describe the relationship between `handspan` and `height`. </span>

```python
# Answer
import seaborn as sns
import matplotlib.pyplot as plt
plt.figure(figsize = (10,8))
sns.regplot(x = 'handspan',
            y = 'height',
            data = body_data)
plt.title('Regplot showing relationship between handspan and height')
plt.xlabel('Handspan')
plt.ylabel('Height')
plt.show()
```

```python
plt.figure(figsize = (10,8))
sns.boxplot(x='handspan',
            y = 'height',
            data =  body_data)
plt.title('Boxplot showing relationship between handspan and height')
plt.xlabel('Handspan')
plt.ylabel('Height')
plt.show()
```

### Problem 21: (5 points)

<span  style="color:purple; font-size:16px">Set the `timestamp` column to be the index. Use the `resample()` method to aggregate the data in 3 hour intervals. Use the `agg()` function to apply the `count` method to the `sex` column. Finally, plot the time-series so we can visualize the number of responses submitted over each 3 hour interval. Describe any pattern that you see in the data.</span>

(Hint: After setting timestamp to be the index, use code like: `data.resample(<stuff>).agg(<stuff>).plot(<stuff>)`)

```python
# Answer
body_data.set_index('timestamp', inplace = True)
resampled_data = body_data.resample('3H').agg({'sex':'count'})
plt.figure(figsize = (10,8))
resampled_data.plot()
plt.xlabel('timestamp')
plt.ylabel('number of responses')
plt.title('Number of responses submitted over 3 hours intervals')
plt.show()
```
