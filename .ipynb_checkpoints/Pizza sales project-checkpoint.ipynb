```python
df_iota = df.copy()
df_iota = df_iota.astype({'order_time':'string'})


df_iota_2 = df_iota.copy()
# df_iota_3 = df_iota_2.explode('split_time')
df_iota_3 = df_iota_2.groupby(['order_id','order_time']).size().reset_index(name='count')
df_iota_3['count'] = 1
df_iota_3['split_time'] = df_iota_3.order_time.str.split(':')
df_iota_3
```

```python
time_df = pd.DataFrame(df_iota_3['split_time'].tolist())

time_df = time_df.rename(columns = {0:'Hour',1:'Minute',2:'Second'})

df_iota_4 = df_iota_3.copy()

df_iota_4 = pd.concat([df_iota_4, time_df], axis=1)
df_iota_4 = df_iota_4.drop('split_time', axis = 1)
df_iota_4.head(10)
```

```python
df_iota_4 = df_iota_4.groupby(['Hour']).count()['count'].sort_values(ascending=False).reset_index(name='count')
df_iota_4
```

<font color = blue>**The most pizza sales is around 12'o' clock which is mostly lunch time.**</font>  </br>

```python
#Plotting above data in figure.
df_iota_5 = df_iota_4.sort_values('Hour',ascending=True).reset_index().drop(columns = 'index')
f, (ax1) = plt.subplots(1, 1, figsize=(10, 6))
x = df_iota_5['Hour']
y5 = df_iota_5['count']

sns.barplot(x=x, y=y5, palette="muted", ax=ax1)

ax1.set_ylabel('Number Sold')
ax1.set_xlabel('Hour of Day (24 hr Scale)')
ax1.axhline(0, color="k", clip_on=False)

plt.title("Number of Pizzas Sold by Hour", fontsize = 16)
plt.tight_layout(h_pad=10)
```
